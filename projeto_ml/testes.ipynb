```python
import pandas as pd

df = pd.read_csv("pizzas.csv")
df
```

```python
df.plot(kind="scatter", x="diametro", y="preco");
```

```python
from sklearn import linear_model

modelo = linear_model.LinearRegression() #modelo da máquina-Regressão Linear
x = df[["diametro"]] #os dois [[]] trazem os dados em formato de data frame
y = df[["preco"]]

modelo.fit(x, y) #treinando o modelo
```

```python
print(modelo.predict([[23]])[0][0]) #modelo realizando uma previsão com base no diametro
```
